```python
import cv2
import matplotlib.pyplot as plt

imgBGR1 = cv2.imread('./sources/apple.png')
imgBGR2 = cv2.imread('./sources/beach.jpg')
imgBGR3 = cv2.imread('./sources/car.jpg')
imgBGR4 = cv2.imread('./sources/dessert.jpg')

imgs = [imgBGR1, imgBGR2, imgBGR3, imgBGR4]
names = ['apple.png', 'beach.jpg', 'car.jpg', 'dessert.jpg']

# 로드 체크 (여기서 FAIL 뜨면 파일이 폴더에 없는 것)
for n, im in zip(names, imgs):
    print(n, "OK" if im is not None else "LOAD FAIL")

# None이면 변환/표시하지 않도록 처리
imgsRGB = [cv2.cvtColor(im, cv2.COLOR_BGR2RGB) if im is not None else None for im in imgs]

fig, ax = plt.subplots(2,2, figsize=(5,5))
fig.suptitle('Sample Photos')  # ✅ Jupyter용 제목 (window title 대신)

pos = [(0,0),(0,1),(1,0),(1,1)]
for (r,c), im, n in zip(pos, imgsRGB, names):
    ax[r][c].axis('off')
    if im is None:
        ax[r][c].set_title("LOAD FAIL")
    else:
        ax[r][c].imshow(im)

plt.subplots_adjust(left=0, right=1, bottom=0, top=1, wspace=0.05, hspace=0.05)
plt.savefig('./sources/sample.jpg', bbox_inches='tight')
plt.show()
```

```python
# 영상의 밝기 조절
src = cv2.imread('./sources/dog.jpg', cv2.IMREAD_GRAYSCALE)    # 원본

if src is None:
    print('Image load failed!')
    sys.exit()

# dst1 = src + 100    # 원본에 100 더한거
# dst2 = cv2.add(src, 100)    # 원본에 100 더하는데 255 넘는 값은 전부 255로

# saturate 연산을 위해 OpenCV의 add() 연산을 사용
# cv2.add(src1, src2, dst=None, mask=None, dtype=None)
# - src1 : (입력) 첫 번째 영상 또는 스칼라 (코드상 src)
# - src2 : (입력) 두 번째 영상 또는 스칼라 (코드상 100)
# - dst : (출력) 덧셈연산의 결과 영상
# - mask : 마스크 영상
# - dtype : 출력 영상(dst)의 타입 (ex. cv2.CV_8U, cv2.CV_32F 등)
dst = cv2.add(src, (100, 100, 100, 0))   # 컬러영상의 밝기 조절(3개채널의 값을 모두 설정)

cv2.imshow('src', src)
# cv2.imshow('dst1', dst1)
# cv2.imshow('dst2', dst2)
cv2.imshow('dst', dst)
cv2.waitKey()

cv2.destroyAllWindows()
```

```python
# 가중치 합 연산을 위해서는 addWeighted() 연산을 사용
# cv2.addWeighted(src1, a, src2, b, gamma, dst=None, dtype=Mpme)
# - src1 : (입력) 첫 번째 영상 또는 스칼라
# - a : 첫 번째 영상 가중치
# - src2 : (입력) 두 번째 영상 또는 스칼라
# - b : 두 번째 영상 가중치
# - gamma : 결과영상에 추가적으로 더할 값
# - dst : (출력) 연산의 결과 영상
# - dtype : 출력영상(dst)의 타입 (ex. cv2.CV_8U, cv2.CV_32F 등)
src1 = cv2.imread('./sources/cat.jpg', cv2.IMREAD_GRAYSCALE)
src2 = cv2.imread('./sources/dog.jpg', cv2.IMREAD_GRAYSCALE)

if src1 is None or src2 is None:
    print('Image load failed!')
    sys.exit()

dst1 = cv2.add(src1, src2)
dst2 = cv2.addWeighted(src1, 0.8, src2, 0.2, 0.0)    # src1 가중치 0.8, src2 가중치 0.2
dst3 = cv2.addWeighted(src1, 0.5, src2, 0.5, 0.0)    # src1 가중치 0.5, src2 가중치 0.5
dst4 = cv2.addWeighted(src1, 0.2, src2, 0.8, 0.0)    # src1 가중치 0.2, src2 가중치 0.8

plt.subplot(231), plt.axis('off'), plt.imshow(src1, 'gray'), plt.title('cat')
plt.subplot(232), plt.axis('off'), plt.imshow(src2, 'gray'), plt.title('dog')
plt.subplot(233), plt.axis('off'), plt.imshow(dst1, 'gray'), plt.title('add')
plt.subplot(234), plt.axis('off'), plt.imshow(dst2, 'gray'), plt.title('0.8_0.2')
plt.subplot(235), plt.axis('off'), plt.imshow(dst3, 'gray'), plt.title('0.5_0.5')
plt.subplot(236), plt.axis('off'), plt.imshow(dst3, 'gray'), plt.title('0.2_0.8')
plt.show()
```

```python
# 뺄셈 연산을 위해서는 subtract() 연산을 사용
# cv2.subtract(src1, src2, dst=None, mask=None, dtype=Mpme)
# - src1 : (입력) 첫 번째 영상 또는 스칼라
# - src2 : (입력) 두 번째 영상 또는 스칼라
# - dst : (출력) 연산의 결과 영상
# - mask : 마스크 영상
# - dtype : 출력영상(dst)의 타입 (ex. cv2.CV_8U, cv2.CV_32F 등)
src1 = cv2.imread('./sources/cat.jpg', cv2.IMREAD_GRAYSCALE)
src2 = cv2.imread('./sources/dog.jpg', cv2.IMREAD_GRAYSCALE)

if src1 is None or src2 is None:
    print('Image load failed!')
    sys.exit()

dst = cv2.subtract(src1, src2)

plt.subplot(131), plt.axis('off'), plt.imshow(src1, 'gray'), plt.title('cat')
plt.subplot(132), plt.axis('off'), plt.imshow(src2, 'gray'), plt.title('dog')
plt.subplot(133), plt.axis('off'), plt.imshow(dst, 'gray'), plt.title('sub')
plt.show()
```

```python
# 차이연산을 위해서는 absdiff() 연산을 사용
# cv2.absdiff(src1, src2, dst=None)
# - src1 : (입력) 첫 번째 영상 또는 스칼라
# - src2 : (입력) 두 번째 영상 또는 스칼라
# - est : (출력) 차이 연산의 결과 영상
src1 = cv2.imread('./sources/abs1.jpg', cv2.IMREAD_GRAYSCALE)
src2 = cv2.imread('./sources/abs2.jpg', cv2.IMREAD_GRAYSCALE)

if src1 is None or src2 is None:
    print('Image load failed!')
    sys.exit()

dst = cv2.absdiff(src1, src2)

plt.subplot(131), plt.axis('off'), plt.imshow(src1, 'gray'), plt.title('abs1')
plt.subplot(132), plt.axis('off'), plt.imshow(src2, 'gray'), plt.title('abs2')
plt.subplot(133), plt.axis('off'), plt.imshow(dst, 'gray'), plt.title('result')
plt.show()
```

```python
# 히스토그램 구하기
# OpenCV에서 영상의 히스토그램을 구하려면 calcHist() 함수를 사용
# cv2.calcHist(images, channels, mask, histSize, ranges, hist=None, accumulate=None)
# - images : 입력 영상 리스트
# - channels : 히스토그램을 구할 채널을 나타내는 리스트
# - mask : 마스크 영상. 입력영상 전체에서 히스토그램을 구하려면 None 지정
# - histSize : 히스토그램 각 차원의 크기(빈(bin)의 개수)를 나타내는 리스트
# - ranges : 히스토그램 각 차원의 최솟값과 최댓값으로 구성된 리스트
# - hist : 계산된 히스토그램 (numpy.ndarray)
# - accumulate : 기존 hist 히스토그램에 누적하려면 True, 새로 만들려면 False
src = cv2.imread('./sources/cat.jpg', cv2.IMREAD_GRAYSCALE)

if src is None:
    print('Image load failed!')
    sys.exit()

hist = cv2.calcHist([src], [0], None, [256], [0, 256])
# src[0](cat.jpg), 256빈, 최솟값0 최댓값256

cv2.imshow('src', src)

plt.plot(hist)
plt.show()

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 컬러영상의 히스토그램
src = cv2.imread('./sources/cat.jpg')

if src is None:
    print('Image load failed!')
    sys.exit()

colors = ['b', 'g', 'r']
bgr_planes = cv2.split(src)

for (p, c) in zip(bgr_planes, colors):
    hist = cv2.calcHist([p], [0], None, [256], [0, 256])
    plt.plot(hist, color=c)

cv2.imshow('src', src)
plt.show()

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 영상의 명암비(Contrast) 조절
# 기본적인 명암비 조절 = np.clip(src, min, max)
# - src : 처리할 배열
# - min : 최소값 설정
# - max : 최대값 설정
import numpy as np

src = cv2.imread('./sources/dog.jpg', cv2.IMREAD_GRAYSCALE)

if src is None:
    print('Image load failed!')
    sys.exit()

alpha = 1.0
dst = np.clip((1+alpha) * src - 128*alpha, 0, 255).astype(np.uint8)

cv2.imshow('src', src)
cv2.imshow('dst', dst)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 히스토그램 스트레칭 : 영상의 히스토그램이 그레이스케일 전 구간에 걸쳐서 나타나도록 변경하는 선형변환기법
# cv2.normalize(src, dst, alpha=None, beta=None, norm_type=None, dtype=None, mask=None)
# - src : 입력 영상
# - dst : 결과 영상 (None 입력)
# - alpha : 목표 노름 값, (원소 값 범위 정규화인 경우) 최솟값
# - beta : (원소 값 범위 정규화인 경우) 최댓값
# - norm_type : 정규화 타입. NORM_INF, NORM_L1, NORM_L2, NORM_MINMAX(히스토그램 스트레칭)
# - dtype : 결과 영상의 타입
# - mask : 마스크 영상
src = cv2.imread('./sources/Hawkes_Bay.jpg', cv2.IMREAD_GRAYSCALE)

dst = cv2.normalize(src, None, 0, 255, cv2.NORM_MINMAX)

cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.waitKey()

cv2.destroyAllWindows()
```

```python
# 히스토그램 평활화 : 특정 그레이스케일 값 근방에서 픽셀 분포가 너무 많이 뭉쳐있는 경우 이를 넓게 펼쳐주는 방식으로 픽셀 값 분포를 조절
# cv2.equalizeHist(src, dst=None)
# - src : 입력 영상
# - dst : 결과 영상 (None 입력)
src = cv2.imread('./sources/Hawkes_Bay.jpg', cv2.IMREAD_GRAYSCALE)

dst = cv2.equalizeHist(src)

cv2.imshow('src', src)
cv2.imshow('dst', dst)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 영상의 이동 변환(Translation transformation)
src = cv2.imread('./sources/scenery.jpg')

aff = np.array([[1, 0, 100], [0, 1, 100]], dtype=np.float32)

dst = cv2.warpAffine(src, aff, (0, 0))

cv2.imshow('src', src)
cv2.imshow('dst', dst)

cv2.waitKey()
cv2.destroyAllWindows()
```
